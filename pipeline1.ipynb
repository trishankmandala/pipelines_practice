```python
import pandas as pd
```

```python
#read csv 
df_transactions = pd.read_csv('transaction_data.csv')
df_customers = pd.read_csv('us_customer_data.csv')
```

```python
#merged two csv files
merged_df = pd.merge(df_transactions,df_customers,on='customer_id')
```

```python
#remove duplicates
merged_df.duplicated().sum()
merged_df.drop_duplicates()
```

```python
#replaced empty email with na values
merged_df['email']=merged_df['email'].replace('',pd.NA)
```

```python
#filled na values with some email
merged_df['email']=merged_df['email'].fillna('unknown@example.com')
```

```python
#filling amount with 0
merged_df['amount'] = merged_df['amount'].fillna(0)
```

```python
def fill_nulls(merged_df):
    for column in merged_df.columns:#strings
        if merged_df[column].dtype=='object':
            merged_df[column]=merged_df[column].fillna(f'unknown_{column}')
        else:
            merged_df[column]=merged_df[column].fillna(0)
    return merged_df
cleaned_csv=fill_nulls(merged_df)
 
```
